```python
import os
base_path = "../input/yoga-pose-image-classification-dataset/dataset"
classes_dir_data = os.listdir(base_path)
num_of_classes = len(classes_dir_data)
print("Total Number of Classes :" , num_of_classes)
class_list = []
number_of_pics_list = []
for x in classes_dir_data:
    dir_of_images = os.path.join("../input/yoga-pose-image-classification-dataset/dataset",x)
    list_of_images = os.listdir(dir_of_images)
    number_of_images = len(list_of_images)
    class_list.append(x)
    number_of_pics_list.append(number_of_images)
```

```python
import pandas as pd
import numpy as np
list_df = pd.DataFrame({
    "Class" : class_list,
    "Number of pictures":number_of_pics_list
})
list_df.sort_values(by = "Number of pictures",ascending = False)
```
