# step 2: preparing the data for modelling
*  encoding:
*      coverting binary vars to 1/0
*      other categorical vars to dummy vars

```python
#yes/no variables
varlist=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
housing[varlist]=housing[varlist].apply(lambda x: x.map({'yes':1,'no':0}))
housing[varlist].head()
```

# dummy variable

```python
#creating dummy variables for furnishing status
dummy= pd.get_dummies(housing['furnishingstatus'],drop_first=True)
dummy
```

```python
housing=housing.drop('furnishingstatus',axis=1)
housing
```

```python
housing=pd.concat([housing,dummy],axis=1)
housing
```

# split into train and test

```python
df_train,df_test=train_test_split(housing,train_size=0.7,random_state=100)
print(df_train.shape)
print(df_test.shape)
```

# rescaling the feature

```python
scaler=MinMaxScaler()
num_vars=['area','bedrooms','bathrooms','stories','parking','price']
df_train[num_vars]=scaler.fit_transform(df_train[num_vars])
df_train.head()
```

# step 3: training the data

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_train.corr(),annot=True)
```

```python
y_train=df_train.pop('price')
x_train=df_train
x_train.head()
```

```python
import statsmodels.api as sm
```

```python
x_train_sm=sm.add_constant(x_train['area'])
lr=sm.OLS(y_train,x_train_sm)
lr_model=lr.fit()
lr_model.params
```

```python
x_train_sm=x_train[['area','bathrooms']]
x_train_sm=sm.add_constant(x_train_sm)
lr=sm.OLS(y_train,x_train_sm)
lr_model=lr.fit()
lr_model.params 
```

```python
lr_model.summary()
```

```python
x_train_sm=x_train[['area','bathrooms','bedrooms']]
x_train_sm=sm.add_constant(x_train_sm)
lr=sm.OLS(y_train,x_train_sm)
lr_model=lr.fit()
lr_model.params 
lr_model.summary()
```

# adding all variables to model

```python
x_train_sm=sm.add_constant(x_train)
lr=sm.OLS(y_train,x_train_sm)
lr_model=lr.fit()
lr_model.params 
lr_model.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['features']=x_train.columns
vif['Vif']=[variance_inflation_factor(x_train.values,i) for i in range(x_train.shape[1])]
vif['Vif']=round(vif['Vif'],2)
vif=vif.sort_values(by="Vif",ascending=False)
vif
```

```python
x=x_train.drop('semi-furnished',axis=1)
x
```

```python
x_train_sm=sm.add_constant(x)
lr=sm.OLS(y_train,x_train_sm)
lr_model=lr.fit()
lr_model.params 
lr_model.summary()
```

```python
vif=pd.DataFrame()
vif['features']=x.columns
vif['Vif']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
vif['Vif']=round(vif['Vif'],2)
vif=vif.sort_values(by="Vif",ascending=False)
vif
```

```python
x=x.drop('bedrooms',axis=1)
```

```python
x_train_sm=sm.add_constant(x)
lr=sm.OLS(y_train,x_train_sm)
lr_model=lr.fit()
lr_model.params 
lr_model.summary()
```

```python
vif=pd.DataFrame()
vif['features']=x.columns
vif['Vif']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
vif['Vif']=round(vif['Vif'],2)
vif=vif.sort_values(by="Vif",ascending=False)
vif
```

# step 4: residual analysis

```python
y_train_pred=lr_model.predict(x_train_sm)
```

```python
res=y_train-y_train_pred
sns.distplot(res)
```
